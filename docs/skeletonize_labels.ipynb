```python
import numpy as np
from napari_segment_blobs_and_things_with_membranes import skeletonize
from skimage.io import imread, imshow
from skimage.filters import gaussian, threshold_otsu
from skimage.measure import label
```

```python
image = imread('https://fiji.sc/samples/blobs.png')
```

```python
blurred = gaussian(image, sigma=10)
binary = blurred > threshold_otsu(blurred)
labels = label(binary)
imshow(labels, cmap="jet")
```

```python
skel = skeletonize(labels)
imshow(skel, cmap="jet")
```

```python
imshow(skel[180:250, 100:150], cmap="jet")
```
